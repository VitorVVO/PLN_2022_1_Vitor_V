```python
import numpy as np
import pandas as pd
import math
```

```python
def fativacao(x,tipo,t):
    if tipo == 'degrau':
        if x >= t:
            y = 1
        else:
            y = 0
    elif tipo == 'sigmoide':
        y = 1/(1 + math.exp(-x))
    elif tipo == 'sinal':
        if x >= 0:
            y = 1
        else:
            y = -1
    else:
        y = 'ERRO: Opcao invalida'
    return y
```

```python
data = [[0,0],[0,1],[1,0],[1,1]]
X = np.array(data)
vet_labels = [0,1,1,0]
labels = np.array(vet_labels)
learning_rate = 0.3
```

# foward

```python
# pesos hidden layer w1 w2 w3 w4
pesos1 = [[0.4,0.6],[0.5,0.3]]

# pesos camada de saida w21 w22
pesos2 = [0.2,0.1]

for i,x in enumerate(X):
  saidas_o=[]
  # hidden layer: sigmoid((x1*w1) + (x2*w3)) e sigmoid((x1*w2) + (x2*w4))
  for j in range(0,len(pesos1)):
    net = x[0]*pesos1[0][j] + x[1]*pesos1[1][j] 
    o = fativacao(net,'sigmoide',0)
    saidas_o.append(o)

  # camada de saida: sigmoid((Oa*w21) + (Ob*w22))
  net = saidas_o[0]*pesos2[0] + saidas_o[1]*pesos2[1] 
  y = fativacao(net,'sigmoide',0)
```

# backpropagation:

```python
# Calculando os erros :

grad_layer1=[]

# para camada de saída: δ = y * (1 − y) * (l(x[i]) − y)
for i,x in enumerate(X):
  grad_layer2 = (y * (1-y)) * (labels[i] - y)

# para as camadas ocultas: δ = y[i] * (1 − y[i]) *  w[i] * δ_out
for neuron in range(0,len(pesos1)):
  grad_layer1.append(saidas_o[neuron] * (1-saidas_o[neuron]) * pesos2[neuron] * grad_layer2)
```

```python
# Reajustando os pesos:

# w[i,j] += lr * δ[i] * y[i,j]
for neuron in range(0,len(pesos1)):
  for peso in range(0,len(pesos1[neuron])):
    pesos1[neuron][peso] += learning_rate*grad_layer1[neuron]*saidas_o[neuron]

# w[i] += lr * δ * y[i]
for peso in range(0,len(pesos2)):
  pesos2[peso] += learning_rate*grad_layer2*y
```

# MLP Completa com backpropagation:

```python
# Inicializando valores importantes
data = [[0,0],[0,1],[1,0],[1,1]]
X = np.array(data)
vet_labels = [0,1,1,0]
labels = np.array(vet_labels)
learning_rate = 0.2
```

```python
# pesos hidden layer w1 w2 w3 w4
pesos1 = [[0.2,0.5],[0.4,0.5]]

# pesos camada de saida w21 w22
pesos2 = [0.3,0.3]

# contador de iterações
it = 0
```

```python
# loop da MLP em 15 iterações
while(it<15):
  print(f'Epoch {it}°:')
  print("------------------------------------------------------------------------------------------------------------")

  grads1=[]
  grads2=[]
  saida_hidden=[]
  saida_output=[]

  for i,x in enumerate(X):
    saidas_o=[]
    print(f'-------Foward {i+1}:-------')
    # hidden layer: sigmoid((x1*w1) + (x2*w3)) e sigmoid((x1*w2) + (x2*w4))
    for j in range(0,len(pesos1)):
      net = x[0]*pesos1[0][j] + x[1]*pesos1[1][j] 
      o = fativacao(net,'sigmoide',0)
      #print('sigmoid('+str(x[0]),'*',pesos1[0][j],'+',x[1],'*',str(pesos1[1][j])+') =',o)
      saidas_o.append(o)
    print("Hidden layer -> Saidas O:",saidas_o)

    # camada de saida: sigmoid((Oa*w21) + (Ob*w22))
    net = saidas_o[0]*pesos2[0] + saidas_o[1]*pesos2[1] 
    y = fativacao(net,'sigmoide',0)
    #print('sigmoid('+str(saidas_o[0]),'*',pesos2[0],'+',saidas_o[1],'*',str(pesos2[1])+') =',y,'\n')
    print("Output layer -> Y:",y)

    print(f'------Backward {i+1}:------')
    # gerando Gradientes
    grad_layer1=[]

    # para camada de saída: δ = y * (1 − y) * (l(x[i]) − y)
    grad_layer2 = (y * (1-y)) * (labels[i] - y)

    # para as camadas ocultas: δ = y[i] * (1 − y[i]) *  w[i] * δ_out
    for neuron in range(0,len(pesos1)):
      grad_layer1.append(saidas_o[neuron] * (1-saidas_o[neuron]) * pesos2[neuron] * grad_layer2)
    print('Hidden layer -> Gradientes:',grad_layer1)
    print('Output layer -> Gradiente:',grad_layer2,'\n')

    grads1.append(grad_layer1)
    grads2.append(grad_layer2)
    saida_hidden.append(saidas_o)
    saida_output.append(y)

  # Reajustando os pesos
  # w[i,j] += lr * δ[i] * y[i]   
  for j in range(0,len(grads1)):
    for neuron in range(0,len(pesos1)):
      for peso in range(0,len(pesos1[neuron])):
        pesos1[neuron][peso] += learning_rate * grads1[j][neuron] * saida_hidden[j][neuron]

  # w[i] += lr * δ * y                         
  for j in range(0,len(grads2)):
    for peso in range(0,len(pesos2)):
      pesos2[peso] += learning_rate * grads2[j] * saida_output[j]

  print("------Pesos Atualizados:------")
  print("Pesos 1:",pesos1)
  print("Pesos 2:",pesos2,"\n")
  print("------------------------------------------------------------------------------------------------------------\n")

  it += 1
```
